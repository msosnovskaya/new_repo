# Немного о функциях, копипасте и хардкодинге

```python
# хотим задать долю какой-то комнаты от площади всех квартиры
data['kitchen_pct'] = data['kitchen_area'] / data['total_area']
data['living_pct'] = data['living_area'] / data['total_area']
```

#### Почему хардкодинг это плохо?
- Когда значения зашиты прямо в коде, внесение изменений становится сложнее. Например, если потребуется изменить одно значение, нужно будет искать все места в коде, где оно используется, и обновлять его везде.
- Это увеличивает вероятность ошибок, поскольку легко пропустить одно из мест или внести ошибку в процесс изменения.
- Трудно конфигурировать. В случае, когда необходимо изменить поведение программы (например, изменить путь к файлу или URL-адрес), потребуется модифицировать сам код, а не просто конфигурационный файл.
- Труднее адаптировать для использования в разных контекстах или проектах, так как в нем жестко зашиты специфические значения.

```python
def pct_cnt(df, name_of_area):
    data[f'{name_of_area}_pct'] = data[f'{name_of_area}_area'] / data['total_area']
    return data
```

```python
data = pct_cnt(df=data, name_of_area='kitchen')
data = pct_cnt(df=data, name_of_area='living')
```

# Результат статистического теста

### Легенда

По окончании учебного года студенты прошли опрос, в котором нужно было оценить курс статистики, который им был дан в последнем семестре. Размах оценки был в пределах от -5 (ужасно) до 5 (фантастически). Курс по статистики шел в онлайн формате для десятков тысяч студентов, поэтому для оценки его успеха мы взяли случайную выборку из 50 оценок.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(123)
lo = np.random.randint(-5, -1, 6)
mid = np.random.randint(0, 3, 38)
hi = np.random.randint(4, 6, 6)
sample = np.append(lo,np.append(mid, hi))
print("Min:" + str(sample.min()))
print("Max:" + str(sample.max()))
print("Mean:" + str(sample.mean()))

plt.hist(sample)
plt.show()
```

Мы начинаем с постановки двух гипотез:

* *Нулевая* гипотеза (**H<sub>0</sub>**) заключается в том, что среднее генеральной совокупности для всех оценок *не* выше 0, и тот факт, что наша выборочная средняя оценка выше 0, обусловлен случайностью выборки.
* *Альтернативная* гипотеза (**H<sub>1</sub>**) заключается в том, что среднее генеральной совокупности действительно выше 0, и тот факт, что наша выборочная средняя оценка выше 0 означает, что наша выборка верно выявила эту тенденцию.

Вы можете записать их как взаимоисключающие выражения вот так:

\begin{equation}H_{0}: \mu \le 0, 
\\ H_{1}: \mu > 0 \end{equation}

В данном случае работаем с односторонней гипотезой 


```python
from scipy.stats import ttest_1samp
```

```python
alpha = 0.05
t, p = ttest_1samp(sample, 0,  alternative='greater')
print('p_value = {:.2}'.format(p))
print('Уровень значимости alpha = {:.2}'.format(alpha))
print('')
if p < alpha:
    print('p_value < alpha, следовательно отвергаем нулевую гипотезу о том, что средняя оценка не выше нуля')
else:
    print('p_value >= alpha, следовательно у нас недостаточно оснований отвергнуть нулевую гипотезу о том, что средняя оценка не выше нуля')

print('')
```

Если нам необходимо проверить двустороннюю гипотезу: 
\begin{equation}H_{0}: \mu = 0, 
\\ H_{1}: \mu != 0 \end{equation}


```python
alpha = 0.05
t,p = ttest_1samp(sample, 0)
print('p_value = {:.2}'.format(p))
print('Уровень значимости alpha = {:.2}'.format(alpha))
print('')
if p < alpha:
    print('p_value < alpha, следовательно отвергаем нулевую гипотезу о равенстве средней оценки нулю')
else:
    print('p_value >= alpha, следовательно у нас недостаточно оснований отвергнуть нулевую гипотезу о равенстве средней оценки нулю')

print('')
```

Вы можете заметить, что удвоение p-value в двустороннем тесте делает его сложнее для отвержения нулевой гипотезы. Это действительно так; нам требуется больше доказательств, потому что мы задаем более сложную гипотезу.

Важно! Если p-value > 0,05, мы не можем отклонить нулевую гипотезу и заключаем, что средняя оценка не превышает 0. **Мы никогда не принимаем нулевую гипотезу, просто нет достаточных доказательств для ее отклонения.**

# Не забываем про красивый вывод переменных и подписи к интерпретации

```python
alpha = 0.05
t,p = ttest_1samp(sample, 0)
if p < alpha:
    print(p, 'Отвергаем')
else:
    print(p, 'Не отвергаем')
```
